# maysics.equation模块使用说明

equation模块包含一个类

|名称|作用|
|---|---|
|Scatter|离散点方程|

equation模块包含一个函数

|名称|作用|
|---|---|
|linear_e|线性方程组|

<br></br>
## 散点组成的方程：Scatter
当散点只有一组时，求与x轴的交点
<br>当散点大于一组时，求各组散点的交点

<br>Scatter(x, y, acc=0.1, kind='linear',xtol=0.1, ytol=0.1)

### DEMO 1-1：一组散点组成的方程求根并绘图

````python
from maysics.equation import Scatter
import numpy as np

x = np.arange(-5, 5, 0.01)
y = x**2 - 2
scatter = Scatter(x, y)
print(scatter.root)
scatter.show()
````

### DEMO 1-2：两组散点组成的方程求根并绘图

````python
from maysics.equation import Scatter
import numpy as np

x1 = np.arange(-5, 5, 0.01)
y1 = x1**2 - 2
x2 = np.arange(-5, 5, 0.01)
y2 = -x2**2 + 2
X = [x1, x2]
Y = [y1, y2]
scatter = Scatter(X, Y)
print(scatter.root)
scatter.show()
````

## 线性方程组：linear_e
linear_e(A, b)
<br>```A```是方程组系数矩阵
<br>```b```是常数项组成的行（列）矩阵

### DEMO 2-1：求解线性方程组$Ax=b$

````python
from maysics.equation import linear_e

A = [[1,2],
     [3,4]]
b = [[5],
     [6]]
linear_e(A, b)
````
